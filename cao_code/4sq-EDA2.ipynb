## Adding "similarity" columns

```python
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
```

```python
# name similarity array;
sim_arr = []
for i in range(len(df)):
    a = df.iloc[[i]].loc[:,'name_1'].values[0]
    b = df.iloc[[i]].loc[:,'name_2'].values[0]
    sim = similar(a,b)
    sim_arr.append(sim)
    if i%5000==0: 
        print('{}; '.format(i), end='')
    if i%50000==0:
        print()
```

```python
df['name_similarity'] = sim_arr
```

```python
# geo-distance array;
dist_arr = []
for i in range(len(df)):
    a = df.iloc[[i]].loc[:,['latitude_1','longitude_1']].values[0]
    b = df.iloc[[i]].loc[:,['latitude_2','longitude_2']].values[0]
    d = np.linalg.norm(a-b)
    dist_arr.append(d)
    if i%5000==0: 
        print('{}; '.format(i), end='')
    if i%50000==0:
        print()
# dist_arr
```

```python
df['geo_distance'] = dist_arr
```
